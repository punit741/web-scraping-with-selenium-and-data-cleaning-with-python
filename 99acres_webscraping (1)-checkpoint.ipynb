# 99Acres.com Web Scraper
This notebook scrapes property listings from 99acres.com for Delhi.
It uses Selenium to control the browser and Pandas to store the data.

```python
# Step 1 - Import all the libraries we need

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import TimeoutException

import time
import re
import numpy as np
import pandas as pd
```

```python
# Step 2 - This function checks if the page has fully loaded
# We fixed 2 bugs from the original:
#   1. 'wait.untill' was a typo, correct spelling is 'until'
#   2. The javascript string was wrong, fixed to: return document.readyState

def wait_for_page_to_load(driver, wait):
    title = driver.title
    try:
        wait.until(
            lambda d: d.execute_script("return document.readyState") == "complete"
        )
    except:
        print("The webpage did NOT fully load:", title)
    else:
        print("The webpage loaded successfully:", title)
```

```python
# Step 3 - Open the Chrome browser with some settings
# These settings make the browser look like a normal user, not a bot

options = webdriver.ChromeOptions()

options.add_argument("--start-maximized")                              # open in full screen
options.add_argument("--disable-blink-features=AutomationControlled")  # hide bot flag
options.add_argument("--incognito")                                    # use incognito mode
options.add_argument("--disable-extensions")                           # no extensions
options.add_argument("--window-size=1920,1080")                        # set screen size
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36")

driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 15)  # wait up to 15 seconds for elements to appear

print("Browser opened successfully")
```

```python
# Step 4 - Open 99acres website

url = "https://www.99acres.com/"
driver.get(url)
wait_for_page_to_load(driver, wait)
```

```python
# Step 5 - Type DELHI in the search bar

try:
    search_bar = wait.until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="keyword2"]'))
    )
    driver.execute_script("arguments[0].scrollIntoView(true);", search_bar)
    search_bar.send_keys("DELHI")
    print("Typed DELHI in search bar")
except:
    print("Could not find the search bar")

time.sleep(3)  # wait for the dropdown suggestions to appear
```

```python
# Step 6 - Click on the Delhi option from the dropdown

try:
    delhi_option = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="1"]'))
    )
except:
    print("Timeout when finding Delhi option")
else:
    delhi_option.click()
    print("Clicked on Delhi")
    time.sleep(1)

# Click the Search button
try:
    search_btn = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="searchform_search_btn"]'))
    )
except:
    print("Timeout when finding Search button")
else:
    search_btn.click()
    print("Clicked Search button")
    time.sleep(3)  # wait for results page to load
```

```python
# Step 7 - Move the budget slider (optional filter)
# We drag the max budget slider to the left to filter by price

try:
    slider = wait.until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="budgetLeftFilter_max_node"]'))
    )
except:
    print("Budget slider not found, skipping")
else:
    actions = ActionChains(driver)
    actions.click_and_hold(slider).move_by_offset(-73, 0).release().perform()
    print("Budget slider moved")
    time.sleep(2)
```

```python
# Step 8 - Scroll through the filter panel
# The filter panel has a right arrow to reveal more filter options
# We keep clicking it until there are no more arrows

while True:
    try:
        arrow = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, "//i[@class='iconS_Common_24 icon_upArrow cc__rightArrow']"))
        )
        arrow.click()
        time.sleep(1)
    except:
        print("Reached end of filter panel")
        break
```

```python
# Step 9 - Apply filters: Verified, Ready to Move, With Photos, With Videos

# Filter 1 - Verified listings
try:
    verified = wait.until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="app"]/div/div/div[4]/div[3]/div[1]/div[3]/section/div/div/div/div/div[1]/div/div[3]'))
    )
except:
    print("Verified filter not found")
else:
    verified.click()
    print("Clicked Verified filter")
    time.sleep(2)

# Filter 2 - Ready to Move
try:
    ready_to_move = wait.until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="app"]/div/div/div[4]/div[3]/div[1]/div[3]/section/div/div/div/div/div[1]/div/div[5]'))
    )
except:
    print("Ready to Move filter not found")
else:
    ready_to_move.click()
    print("Clicked Ready to Move filter")
    time.sleep(2)

# Filter 3 - With Photos
try:
    photos_btn = wait.until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="app"]/div/div/div[4]/div[3]/div[1]/div[3]/section/div/div/div/div/div[2]/div/div[6]'))
    )
except:
    print("With Photos filter not found")
else:
    photos_btn.click()
    print("Clicked With Photos filter")
    time.sleep(2)

# Filter 4 - With Videos
try:
    videos_btn = wait.until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="app"]/div/div/div[4]/div[3]/div[1]/div[3]/section/div/div/div/div/div[2]/div/div[7]'))
    )
except:
    print("With Videos filter not found")
else:
    videos_btn.click()
    print("Clicked With Videos filter")
    time.sleep(3)  # give page time to reload with filters applied
```

```python
# Step 10 - Scrape all pages
# We go through each page, grab the listing details, then click Next

page_count = 0
data = []  # this list will hold all the property info

while True:
    page_count += 1
    print("Scraping page", page_count, "...")

    # Wait for the listing cards to appear on this page
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "tupleNew__contentWrap"))
        )
    except:
        print("Listings did not load in time.")
        break

    # Get all listing cards on this page
    rows = driver.find_elements(By.CLASS_NAME, "tupleNew__contentWrap")

    if not rows:
        print("No listings found on this page.")
        break

    # Loop through each listing and grab the details
    for row in rows:

        # Property name
        try:
            name = row.find_element(By.CLASS_NAME, "tupleNew__headingNrera").text
        except:
            name = np.nan

        # Property type and location (e.g. '3 BHK Flat in Lajpat Nagar, Delhi')
        try:
            location = row.find_element(By.CLASS_NAME, "tupleNew__propType").text
        except:
            location = np.nan

        # Price
        try:
            price = row.find_element(By.CLASS_NAME, "tupleNew__priceValWrap").text
        except:
            price = np.nan

        # Area and BHK (these two come from the same class, so we grab both together)
        try:
            elements = row.find_elements(By.CLASS_NAME, "tupleNew__area1Type")
            if len(elements) == 2:
                area = elements[0].text
                bhk  = elements[1].text
            else:
                area = np.nan
                bhk  = np.nan
        except:
            area = np.nan
            bhk  = np.nan

        # Save this listing
        data.append({
            "name"     : name,
            "location" : location,
            "price"    : price,
            "area"     : area,
            "bhk"      : bhk
        })

    print("  Properties collected so far:", len(data))

    # Click the Next button to go to the next page
    try:
        next_button = WebDriverWait(driver, 15).until(
            EC.element_to_be_clickable((By.XPATH, '//a[contains(text(),"Next")]'))
        )
        driver.execute_script("arguments[0].click();", next_button)

        # Wait for the page to change (old Next button disappears)
        WebDriverWait(driver, 15).until(EC.staleness_of(next_button))
        time.sleep(4)  # small wait for the new page to load

    except:
        print("Reached the last page.")
        break

print("")
print("Total pages visited:", page_count)
print("Total properties scraped:", len(data))
```

```python
# Step 11 - Put all the scraped data into a DataFrame

df = pd.DataFrame(data)

print("Shape of raw data:", df.shape)
print("")
print("How many null values in each column:")
print(df.isnull().sum())

df.head(10)
```

## Data Cleaning
The raw scraped data has some issues we need to fix:
- Some property names have ratings attached like `Godrej South Estate\n4.0`
- Price is a string like `Rs.2.8 Cr` — we want a number
- Area is a string like `1,260 sqft` — we want a number
- BHK is a string like `3 BHK` — we want just the number 3
- Some rows are exact duplicates

```python
# Cleaning Step 1 - Remove duplicate rows

rows_before = len(df)
df = df.drop_duplicates()
df = df.reset_index(drop=True)

print("Rows removed as duplicates:", rows_before - len(df))
print("Rows remaining:", len(df))
```

```python
# Cleaning Step 2 - Fix the name column
# Some names look like: 'Godrej South Estate\n4.0'
# We want to remove the '\n4.0' part and extra spaces

df["name"] = df["name"].astype(str)
df["name"] = df["name"].str.replace(r"\\n[\d.]+", "", regex=True)  # remove ratings like \n4.0
df["name"] = df["name"].str.replace("\n", " ")                      # replace newlines with space
df["name"] = df["name"].str.strip()                                  # remove extra spaces
df["name"] = df["name"].replace("nan", np.nan)                       # put back proper nulls

print("Name column cleaned")
print(df["name"].head(5))
```

```python
# Cleaning Step 3 - Convert price to a number
# Raw examples: 'Rs.2.8 Cr', 'Rs.85 L'
# We convert everything to Crore (1 Crore = 100 Lakh)

def convert_price(price_text):
    if pd.isna(price_text):
        return np.nan

    text = str(price_text)
    text = text.replace("Rs.", "").replace("Rs", "").replace(",", "").strip()

    # find the number and the unit (Cr or L)
    match = re.search(r"([\d.]+)\s*(Cr|L)?", text, re.IGNORECASE)
    if not match:
        return np.nan

    number = float(match.group(1))
    unit   = str(match.group(2)).upper() if match.group(2) else ""

    if unit == "L":    # convert Lakh to Crore
        return round(number / 100, 4)
    else:              # already in Crore
        return number

df["price_cr"] = df["price"].apply(convert_price)

print("Price column converted")
print(df[["price", "price_cr"]].head(5))
```

```python
# Cleaning Step 4 - Convert area to a number
# Raw examples: '1,260 sqft', '23,895 sqft'
# We want just the number: 1260.0

def convert_area(area_text):
    if pd.isna(area_text):
        return np.nan
    text  = str(area_text).replace(",", "")  # remove commas
    match = re.search(r"([\d.]+)", text)
    return float(match.group(1)) if match else np.nan

df["area_sqft"] = df["area"].apply(convert_area)

print("Area column converted")
print(df[["area", "area_sqft"]].head(5))
```

```python
# Cleaning Step 5 - Convert BHK to a number
# Raw examples: '3 BHK', '2 BHK'
# We want just the number: 3, 2

df["bhk_count"] = df["bhk"].astype(str).str.extract(r"(\d+)").astype(float)

print("BHK column converted")
print(df[["bhk", "bhk_count"]].head(5))
```

```python
# Cleaning Step 6 - Add a new column: price per sqft
# This tells us how expensive a property is per square foot
# Formula: (price in crore x 1,00,00,000) / area in sqft

df["price_per_sqft"] = ((df["price_cr"] * 10000000) / df["area_sqft"]).round(0)

print("Price per sqft column added")
```

```python
# Cleaning Step 7 - Drop rows that have no useful numeric data at all

rows_before = len(df)
df = df.dropna(subset=["price_cr", "area_sqft", "bhk_count"], how="all")
df = df.reset_index(drop=True)

print("Rows removed (no numeric data):", rows_before - len(df))
print("Final row count:", len(df))
```

```python
# Step 12 - View the final cleaned DataFrame

print("Final shape:", df.shape)
print("")
print("Null values per column:")
print(df.isnull().sum())

df.head(10)
```

```python
# Step 13 - Print a simple summary of what we scraped

print("Total properties:", len(df))
print("BHK types available:", sorted(df["bhk_count"].dropna().unique()))
print("Price range: Rs.", df["price_cr"].min(), "Cr  to  Rs.", df["price_cr"].max(), "Cr")
print("Average price: Rs.", round(df["price_cr"].mean(), 2), "Cr")
print("Average area:", round(df["area_sqft"].mean(), 0), "sqft")
print("")
print("Count by BHK type:")
print(df["bhk_count"].value_counts().sort_index())
```

```python
# Step 14 - Save the data to a CSV file
# You can open this file in Excel or upload it to GitHub

df.to_csv("99acres_delhi_properties.csv", index=False, encoding="utf-8-sig")

print("File saved: 99acres_delhi_properties.csv")
print("Rows:", len(df))
print("Columns:", list(df.columns))
```

```python
# Step 15 - Close the browser

driver.quit()
print("Browser closed")
```
